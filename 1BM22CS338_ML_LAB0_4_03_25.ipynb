ANALYZING SALES DATA

```python
# Reading sales data from a CSV file
sales_df = pd.read_csv("sample_sales_data.csv")
# Displaying the first few rows of the dataset
print("First few rows of the sales data:")
print(sales_df.head())
```

```python
# Grouping by Region and calculating total sales
sales_by_region = sales_df.groupby('Region')['Sales'].sum()
print("\nTotal sales by region:")
print(sales_by_region)
```

```python
# Grouping by Product and calculating total quantity sold
best_selling_products = sales_df.groupby('Product')['Quantity'].sum().sort_values(ascending=False)
print("\nBest-selling products by quantity:")
print(best_selling_products)
```

```python
# Saving the sales by region data to a CSV file
sales_by_region.to_csv('sales_by_region.csv')
# Saving the best-selling products data to a CSV file
best_selling_products.to_csv('best_selling_products.csv')
print("\nAnalysis results saved to CSV files.")
```

IMPORTING STOCK MARKET DATA FOR ANALYSIS

In this real-life scenario, we’ll use the Yahoo Finance API to download historical stock market data for “Reliance Industries Ltd., Tata Consultancy Services Ltd., Infosys Ltd.” and analyze it with Pandas.

```python
# Step 1: Import required libraries
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 2: Downloading Stock Market Data
# Define the ticker symbols for Indian companies
tickers = ["RELIANCE.NS", "TCS.NS", "INFY.NS"]

# Fetch historical data for the last 1 year
data = yf.download(tickers, start="2022-10-01", end="2023-10-01", group_by='ticker')

# Display the first 5 rows of the dataset
print("First 5 rows of the dataset:")
print(data.head())
```

```python
# Step 3: Basic Data Exploration
# Check the shape of the dataset
print("\nShape of the dataset:")
print(data.shape)

# Check column names
print("\nColumn names:")
print(data.columns)

# Summary statistics for a specific stock (e.g., Reliance)
reliance_data = data['RELIANCE.NS'].copy()  # Use copy() to prevent SettingWithCopyWarning
print("\nSummary statistics for Reliance Industries:")
print(reliance_data.describe())

# Calculate daily returns
reliance_data['Daily Return'] = reliance_data['Close'].pct_change()

# Display the first 5 rows with daily returns
print("\nFirst 5 rows with Daily Return:")
print(reliance_data.head())
```

```python
# Step 4: Plot the Closing Price of Reliance Industries
import matplotlib.pyplot as plt

# Plot the Closing Price and Daily Returns
plt.figure(figsize=(12, 6))

# **Plot 1: Closing Price**
plt.subplot(2, 1, 1)
plt.plot(reliance_data.index, reliance_data['Close'], color='blue', label="Closing Price")
plt.title("Reliance Industries - Closing Price")
plt.ylabel("Price (INR)")
plt.grid(True)
plt.legend()

# **Plot 2: Daily Returns**
plt.subplot(2, 1, 2)
plt.plot(reliance_data.index, reliance_data['Daily Return'], color='orange', label="Daily Return")
plt.title("Reliance Industries - Daily Returns")
plt.xlabel("Date")
plt.ylabel("Return (%)")
plt.grid(True)
plt.legend()

# **Adjust layout & show plot**
plt.suptitle("Reliance Industries - Stock Analysis", fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Step 5: Saving the Processed Data to a New CSV File
# Save the Reliance data to a CSV file
reliance_data.to_csv('reliance_stock_data.csv')
print("\nReliance stock data saved to 'reliance_stock_data.csv'.")
```

```python
# Step 1: Import required libraries
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Define the stock tickers & date range
tickers = ["HDFCBANK.NS", "ICICIBANK.NS", "KOTAKBANK.NS"]
start_date = "2024-01-01"
end_date = "2024-12-30"

# Step 3: Download historical stock data
data = yf.download(tickers, start=start_date, end=end_date, group_by='ticker')

# Step 4: Extract closing prices
closing_prices = data.xs(key="Close", axis=1, level=1)

# Step 5: Calculate daily returns
daily_returns = closing_prices.pct_change()

# Step 6: Plot Closing Prices for all three banks
plt.figure(figsize=(12, 6))
for ticker in tickers:
    plt.plot(closing_prices.index, closing_prices[ticker], label=ticker)

plt.title("Closing Prices of HDFC, ICICI, and Kotak Mahindra Banks (2024)")
plt.xlabel("Date")
plt.ylabel("Stock Price (INR)")
plt.legend()
plt.grid(True)
plt.show()

# Step 7: Plot Daily Returns for each bank on separate graphs
for ticker in tickers:
    plt.figure(figsize=(10, 4))
    plt.plot(daily_returns.index, daily_returns[ticker], label=f"{ticker} Daily Return", color="orange")
    plt.title(f"Daily Returns - {ticker}")
    plt.xlabel("Date")
    plt.ylabel("Daily Return (%)")
    plt.legend()
    plt.grid(True)
    plt.show()
```
